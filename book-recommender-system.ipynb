```python
ratings = pd.read_csv('BX-Book-Ratings.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
```

```python
ratings.head(3)
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'book_rating'},inplace=True)
```

```python
ratings.head(3)
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head(3)
```

```python
ratings_with_books = ratings.merge(books,on = 'ISBN')
```

```python
ratings_with_books.head()
```

```python
num_rating = ratings_with_books.groupby('title')['book_rating'].count().reset_index()
```

```python
num_rating
```

```python
num_rating
```

```python
num_rating.rename(columns={
    'book_rating':'no_of_rating'},inplace=True)
```

```python
num_rating
```

```python
final_rating = ratings_with_books.merge(num_rating,on='title')
```

```python
final_rating.head(3)
```

```python
final_rating = final_rating[final_rating['no_of_rating']>50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['title','user_id'],inplace=True)
```

```python
final_rating
```

```python
arr = final_rating['title'].unique()
pass1 = pd.DataFrame(arr)
pass1
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='book_rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance,suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors = 6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
            
```

```python
recommend_book('Year of Wonders')
```
